```python
#Gonna import the atlas, process it, display it, take its laplacian, etc. 
#%matplotlib inline
import graph
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
coords = np.asarray([[float(cord) for cord in re.split("\n | ", line)] for line in open("1000_coords_3column.txt")])
dist, idx = graph.distance_scipy_spatial(coords, k=100, metric="euclidean")
A = graph.adjacency(dist, idx).astype(np.float)
```

```python
def display_single(sample):
    plt.imshow(sample)
    plt.gray()
    plt.show()
    
display_single(A.toarray())
```
